```python
# Import training data
df_train = pd.read_csv('./dataset/data_set_ALL_AML_train.csv')
print(df_train.shape)
df_train.head()
```

```python
# Import test data
df_test = pd.read_csv('./dataset/data_set_ALL_AML_independent.csv')
print(df_test.shape)
df_test.head()
```

```python
# Removing the call columns
train_to_keep = [col for col in df_train.columns if 'call' not in col]
test_to_keep = [col for col in df_test.columns if 'call' not in col]

X_train = df_train[train_to_keep]
X_test = df_test[test_to_keep]

# Taking transpose
X_train = X_train.T
X_test = X_test.T

X_test.head()
```

```python


# Remove first row and set the second row as columns
X_train.columns = X_train.iloc[1]
X_train = X_train.iloc[2:]
X_test.columns = X_test.iloc[1]
X_test = X_test.iloc[2:]

# Setting Index Name
X_train.index.name = "Patient"
X_test.index.name = "Patient"

# Converting the index to int
X_train.index = X_train.index.astype(int)
X_test.index = X_test.index.astype(int)

# Converting the data to int
X_train = X_train.astype(int)
X_test = X_test.astype(int)

# Sort the dataframe
X_train = X_train.sort_index()
X_test = X_test.sort_index()

X_train.head()
```

```python
# Save the dataframes
X_train.to_csv('./cleaned_dataset/X_train.csv')
X_test.to_csv('./cleaned_dataset/X_test.csv')
```

```python
# Merge the two dataframes
X = pd.concat([X_train, X_test])
print(X.shape)

# Save the merged dataframe
X.to_csv('./cleaned_dataset/merged_dataset.csv')

X.head()
```
